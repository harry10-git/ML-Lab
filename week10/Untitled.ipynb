1. Use the IRIS dataset, implement the SVM classifier in python (make use of scikit-learn library), to do the
following.
a. Apply the kernel functions such as linear, polynomial, Radial basis functions and Sigmoid.
b. Plot the scatter plot of the input features.
c. Plot the decision boundary.

```python
import matplotlib.pyplot as plt
from sklearn.inspection import DecisionBoundaryDisplay
from sklearn.svm import SVC
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
# Store features matrix in X
X= iris.data[:,:2]#Store target vector in 
y= iris.target
```

```python
print(iris.feature_names)
```

```python
svm = SVC(kernel="rbf", gamma=0.5, C=1.0)
```

```python
svm.fit(X, y)
DecisionBoundaryDisplay.from_estimator(
svm,
X,
response_method="predict",
cmap=plt.cm.Spectral,
alpha=0.8,
xlabel=iris.feature_names[0],
ylabel=iris.feature_names[1],
)
plt.scatter(X[:, 0], X[:, 1],
c=y,
s=20, edgecolors="k")
plt.show()
```

```python
svm = SVC(kernel="linear", gamma=0.5, C=1.0)
svm.fit(X, y)
DecisionBoundaryDisplay.from_estimator(
svm,
X,
response_method="predict",
cmap=plt.cm.Spectral,
alpha=0.8,
xlabel=iris.feature_names[0],
ylabel=iris.feature_names[1],
)
plt.scatter(X[:, 0], X[:, 1],
c=y,
s=20, edgecolors="k")
plt.show()
```

```python
svm = SVC(kernel="poly", gamma=0.5, C=1.0)
svm.fit(X, y)
DecisionBoundaryDisplay.from_estimator(
svm,
X,
response_method="predict",
cmap=plt.cm.Spectral,
alpha=0.8,
xlabel=iris.feature_names[0],
ylabel=iris.feature_names[1],
)
plt.scatter(X[:, 0], X[:, 1],
c=y,
s=20, edgecolors="k")
plt.show()
```

```python
svm = SVC(kernel="sigmoid", gamma=0.5, C=1.0)
svm.fit(X, y)
DecisionBoundaryDisplay.from_estimator(
svm,
X,
response_method="predict",
cmap=plt.cm.Spectral,
alpha=0.8,
xlabel=iris.feature_names[0],
ylabel=iris.feature_names[1],
)
plt.scatter(X[:, 0], X[:, 1],
c=y,
s=20, edgecolors="k")
plt.show()
```
