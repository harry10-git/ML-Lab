```python
import matplotlib.pyplot as plt
```

```python
# Line plot
x = [5, 2, 9, 4, 7]
y = [10, 5, 8, 4, 2]
plt.plot(x,y)
plt.show()
```

```python
# Bar Plot
x = [5, 2, 9, 4, 7]
y = [10, 5, 8, 4, 2]
plt.bar(x,y)
plt.show()
```

```python
#numPy to randomly generate an array with 250 values, 
#where the values will concentrate around 170, and the standard deviation is 10. 
import numpy as np
x = np.random.normal(170,10,250)
plt.hist(x)
plt.show()
```
